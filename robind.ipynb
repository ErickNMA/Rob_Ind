```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots

#Exibição dos gráficos em janela externa:
#%matplotlib qt

# Configurações de plot:
plt.style.use([
    'grid',
    'retro'
])
#plt.rcParams['lines.linewidth'] = 3
#plt.rcParams['font.size'] = 20
plt.rcParams['figure.figsize'] = (12, 8)
```

```python
#Vetor de tempo:
t0 = 0
tf = 8
ts = 1e-2
t = np.arange(t0, tf, ts)
```

```python
#Para a junta 1:
q1 = ((1.875*(t**2)) - (0.15625*(t**3)))
q1d = ((3.75*t) - (0.46875*(t**2)))
q1dd = (3.75 - (0.9375*t))
```

```python
plt.figure(1)
plt.subplots(3, 1, sharex='all')
plt.subplot(3, 1, 1)
plt.plot(t, q1, color='C1')
plt.ylabel('$q_1(t)~~[°]$')
plt.subplot(3, 1, 2)
plt.plot(t, q1d, color='C2')
plt.ylabel('$\dot{q}_1(t)~~[°/s]$')
plt.subplot(3, 1, 3)
plt.plot(t, q1dd, color='C4')
plt.ylabel('$\ddot{q}_1(t)~~[°/s²]$')
plt.xlabel('Tempo [s]')
plt.savefig('4b-j1.png', dpi=600, transparent=True, bbox_inches='tight')
```

```python
#Para a junta 2:

#Índices de corte:
i1 = int(2/ts)
i2 = int(6/ts)

#Aceleração:
q2dd = np.block([(5*np.ones(i1)), np.zeros(i2-i1), (-5*np.ones(i1))])
#Velocidade:
q2d = np.block([(5*t)[0:i1+1], (10*np.ones(i2-i1)), (40-(5*t))[i2+1:]])
#Posição:
q2 = np.block([(2.5*(t**2))[0:i1+1], (-10+(10*t))[i1+1:i2+1], (-100+(40*t)-(2.5*(t**2)))[i2+1:]])
```

```python
plt.figure(2)
plt.subplots(3, 1, sharex='all')
plt.subplot(3, 1, 1)
plt.plot(t, q2, color='C1')
plt.ylabel('$q_2(t)~~[°]$')
plt.subplot(3, 1, 2)
plt.plot(t, q2d, color='C2')
plt.ylabel('$\dot{q}_2(t)~~[°/s]$')
plt.subplot(3, 1, 3)
plt.plot(t, q2dd, color='C4')
plt.ylabel('$\ddot{q}_2(t)~~[°/s²]$')
plt.xlabel('Tempo [s]')
plt.savefig('4b-j2.png', dpi=600, transparent=True, bbox_inches='tight')
```

```python
#LSPB -> Cúbica:

#Índices de corte:
i1 = int(1/ts)
i2 = int(4/ts)
i3 = int(5/ts)

#Aceleração:
q2dd = np.block([(25*np.ones(i1)), np.zeros(i2-i1), (-25*np.ones(i1)), (-50.4+(6.72*t))[i3:]])
#Velocidade:
q2d = np.block([(25*t)[0:i1+1], (25*np.ones(i2-i1)), (125-(25*t))[i2+1:i3], (168-(50.4*t)+(3.36*(t**2)))[i3:]])
#Posição:
q2 = np.block([(-20+(12.5*(t**2)))[0:i1+1], (-32.5+(25*t))[i1+1:i2+1], (-232.5+(125*t)-(12.5*(t**2)))[i2+1:i3], (-270+(168*t)-(25.2*(t**2))+(1.12*(t**3)))[i3:]])
```

```python
plt.figure(3)
plt.subplots(3, 1, sharex='all')
plt.subplot(3, 1, 1)
plt.plot(t, q2, color='C1')
plt.ylabel('$q_2(t)~~[°]$')
plt.subplot(3, 1, 2)
plt.plot(t, q2d, color='C2')
plt.ylabel('$\dot{q}_2(t)~~[°/s]$')
plt.subplot(3, 1, 3)
plt.plot(t, q2dd, color='C4')
plt.ylabel('$\ddot{q}_2(t)~~[°/s²]$')
plt.xlabel('Tempo [s]')
plt.savefig('5-b.png', dpi=600, transparent=True, bbox_inches='tight')
```
